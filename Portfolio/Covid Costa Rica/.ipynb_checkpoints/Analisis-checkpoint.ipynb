## Importando los Módulos y los Datos

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('primera2.xlsx')
df.head()
```

## Limpiando el DataFrame

```python
df2 = df

try:
    #Eliminando Columnas basura
    del df2['Unnamed: 13']
    del df2['Unnamed: 18']
    del df2['DV1 - PRIMER DOSIS VIDA']
    del df2['DV2 - SEGUNDA DOSIS VIDA']
    df2 = df2.drop(labels=4, axis=0)

    #Renombrando Columnas
    df2 = df2.rename(columns = {'Etiquetas de fila':'Mes'})
    display(df2.head(20))
except:
    print("Corra nuevamente la celda :314")
```

***
## Datos de Vacunación Costa Rica - Divido por dosis
***

### Obteniendo los nombres de las provincias en las columnas

```python
columna_provincias = df2.columns[1:]
provincias = [i.replace('DV1 - ', '').replace('DV2 - ', '') for i in columna_provincias]
provincias = list(set(provincias))
```

### Obteniendo los datos de las dosis por provincia para generar los gráficos y los totales

```python
try:
    def datos_por_dosis(provincias):
        for i in provincias:
            
            #grafico primera dosis
            plt.plot(df2.Mes,df2[[f'DV1 - {i}']], alpha = 0.75, label = "DV1",markersize=5,linewidth=1)
            
            #grafico segunda dosis
            plt.plot(df2.Mes,(df2[[f'DV2 - {i}']]), alpha = 0.75, label = "DV2",markersize=5,linewidth=1)         

            #Diseño del grafico
            plt.xlabel("Mes")
            plt.ylabel("Cantidad")
            plt.title(i)
            
            plt.show()

except:
    print("Ha ocurrido un error")
```

***
## Visualización de Datos - Por Dosis
***

```python
datos_por_dosis(provincias)
```

***
## Datos de Vacunación Costa Rica - Totales
***

```python
try:
    def totales(provincias):
        for i in provincias:
            
            #grafico totales
            plt.plot(i,df2[f'DV1 - {i}'].sum() + df2[f'DV2 - {i}'].sum())
            
            #Diseño del grafico
            plt.xlabel("Provincia")
            plt.ylabel("Cantidad")
            
        plt.show()

except:
    print("Ha ocurrido un error")
```

```python
totales(provincias)
```
